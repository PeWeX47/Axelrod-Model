```python
from AxelrodModel import AxelrodModel

import matplotlib.pyplot as plt
```

```python
# utworzenie instancji klasy AxelrodModel
model = AxelrodModel(N = 100,
                    width = 10, 
                    height = 10, 
                    features = 10, 
                    traits = 240
                    )

# liczba iteracji modelu
for i in range(400000):
    model.step()

data = model.datacollector.get_model_vars_dataframe()
data.plot()
```

```python
# wyświetlenie końcowego stanu symulacji
plt.imshow(model.get_color_matrix())
plt.show()
```
